## Part 1: Preprocessing

```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import numpy as np
from tensorflow.keras.models import Model
from tensorflow.keras import layers

#  Import and read the attrition data
attrition_df = pd.read_csv('https://static.bc-edx.com/ai/ail-v-1-0/m19/lms/datasets/attrition.csv')
attrition_df.head()
```

```python
# Determine the number of unique values in each column.
attrition_df.nunique()
```

```python
# Create y_df with the Attrition and Department columns
y_df = attrition_df[['Attrition', 'Department']]
y_df.head()
```

```python
# Create a list of at least 10 column names to use as X data
selected_columns = ['Age', 'BusinessTravel', 'DistanceFromHome', 'Education',
                    'EnvironmentSatisfaction', 'JobSatisfaction', 'NumCompaniesWorked',
                    'OverTime', 'WorkLifeBalance', 'HourlyRate']


# Create X_df using your selected columns
X_df = attrition_df[selected_columns]

# Show the data types for X_df
X_df.dtypes
```

```python
# Split the data into training and testing sets
from sklearn.model_selection import train_test_split

```

```python
# Convert your X data to numeric data types however you see fit
# Add new code cells as necessary
X_df_encoded = pd.get_dummies(X_df)
X_train, X_test, y_train, y_test = train_test_split(X_df_encoded, y_df, test_size=0.2, random_state=42)
X_train.head()
```

```python
# Create a StandardScaler
scaler = StandardScaler()

# Fit the StandardScaler to the training data
X_train_scaled = scaler.fit_transform(X_train)

# Scale the training and testing data
X_test_scaled = scaler.transform(X_test)
X_train_scaled[:5]
```

```python
# Create a OneHotEncoder for the Department column
from sklearn.preprocessing import OneHotEncoder
department_encoder = OneHotEncoder(sparse_output=False)

# Fit the encoder to the training data
y_train_department_encoded = department_encoder.fit_transform(y_train[['Department']])
y_test_department_encoded = department_encoder.transform(y_test[['Department']])

# Create two new variables by applying the encoder
# to the training and testing data
y_train_department_encoded[:5]

```

```python
# Create a OneHotEncoder for the Attrition column
attrition_encoder = OneHotEncoder(sparse_output=False)

# Fit the encoder to the training data
y_train_attrition_encoded = attrition_encoder.fit_transform(y_train[['Attrition']])
y_test_attrition_encoded = attrition_encoder.transform(y_test[['Attrition']])

# Create two new variables by applying the encoder
# to the training and testing data
y_train_attrition_encoded[:5]
```

## Create, Compile, and Train the Model

```python
# Find the number of columns in the X training data
input_dim = X_train_scaled.shape[1]
print(f"Number of columns in the X training data: {input_dim}")

# Create the input layer
input_layer = layers.Input(shape=(input_dim,))

# Create at least two shared layers
shared_layer_1 = layers.Dense(64, activation='relu')(input_layer)
shared_layer_2 = layers.Dense(32, activation='relu')(shared_layer_1)
```

```python
# Create a branch for Department
# with a hidden layer and an output layer

# Create the hidden layer
department_hidden_layer = layers.Dense(16, activation='relu')(shared_layer_2)

# Create the output layer
department_output_layer = layers.Dense(y_train_department_encoded.shape[1], activation='softmax')(department_hidden_layer)
```

```python
# Create a branch for Attrition
# with a hidden layer and an output layer

# Create the hidden layer
attrition_hidden_layer = layers.Dense(16, activation='relu')(shared_layer_2)

# Create the output layer
attrition_output_layer = layers.Dense(y_train_attrition_encoded.shape[1], activation='softmax')(attrition_hidden_layer)
```

```python
# Create the model
model = Model(inputs=input_layer, outputs=[department_output_layer, attrition_output_layer])


# Compile the model
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

# Summarize the model
model.summary()
```

```python
# Train the model
model.compile(optimizer='adam',
              loss=['categorical_crossentropy', 'categorical_crossentropy'],
              metrics={'dense_3': 'accuracy', 'dense_5': 'accuracy'})

history = model.fit(
    X_train_scaled,
    [y_train_department_encoded, y_train_attrition_encoded],
    epochs=50,
    batch_size=32,
    validation_data=(X_test_scaled, [y_test_department_encoded, y_test_attrition_encoded])
)
```

```python
# Evaluate the model with the testing data
evaluation = model.evaluate(X_test_scaled, [y_test_department_encoded, y_test_attrition_encoded])
print(f"Loss and accuracy for Department: {evaluation[1]}")
print(f"Loss and accuracy for Attrition: {evaluation[2]}")
```

```python
# Print the accuracy for both department and attrition
evaluation = model.evaluate(X_test_scaled, [y_test_department_encoded, y_test_attrition_encoded])
print(evaluation)
```

1. Accuracy might not always be the best metric for this kind of problem. If there’s an imbalance in the data, like way more people staying with the company than leaving, then accuracy can be misleading. For example, if 90% of employees don’t leave, the model could guess that nobody ever leaves and still be right 90% of the time—this would give a high accuracy but wouldn’t be helpful in identifying the employees who actually leave. In these cases, metrics like precision (how many of the ones predicted to leave actually do) and recall (how many of the ones who do leave were predicted) are better because they give more details about the model’s performance for both classes (those who leave and those who stay).
2. I used the softmax activation function for both the Department and Attrition outputs. Softmax is great when we’re working with classification problems that involve more than two possible answers, like when we’re trying to predict which department an employee belongs to or whether someone will stay or leave. Softmax turns the outputs into probabilities, making sure the values add up to 1, so the model can say, for example, “I’m 70% sure this person belongs in Sales, 20% sure it’s R&D, and 10% sure it’s HR.” It’s really helpful for multi-class problems like this one!
3. There are a few ways we could make this model better:

Tuning the model: We could experiment with different setups for the model, like adjusting the number of layers or neurons, changing how fast the model learns, or testing out different batch sizes. This is called hyperparameter tuning, and it can really help improve the model’s performance.
Dealing with imbalanced data: If there are way more people staying with the company than leaving, we could use methods to help the model pay more attention to the less common group (those who leave), like oversampling (creating more examples of the minority group) or giving the minority class more weight during training.
Preventing overfitting: To make sure the model doesn’t just memorize the training data, we could add things like dropout layers (which randomly turn off some neurons during training) or regularization (which makes sure the model doesn’t get too complex).
Adding more data: We could include more useful features or try creating new ones from the existing data (like combining columns or transforming them into something more useful). More or better data can often lead to a better model.
